```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")

plt.xlabel("x-axis") # making x-axis
plt.ylabel("y-axis") #making y-axis

xpoints=np.array([0,5]) #making array of x point values
ypoints=np.array([10,60]) #making array of y point values

plt.plot(xpoints,ypoints)

# to display the line plot we use show()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.title("temparature vs presuure")
plt.xlabel("temparature")
plt.ylabel("pressure")

temparature=np.array([10,50])  #xAXIS
pressure=np.array([100,150])  #Y AXIS

plt.plot(temparature,pressure) #ploting graph

#displaying graph
plt.show()
```

```python
# MULTIPLE POINTS IN SINGLE ARRAY
import matplotlib .pyplot as plt
import numpy as np
plt.title("dist vs speed")
dist=np.array([10,3,4,5]) #x axis

speed=np.array([40,45,37,50]) # y axis

plt.plot(dist,speed)

plt.show()
```

```python
# adding aditional parameers
plt.plot(dist,speed,marker='o',color='r')

plt.show()
```

```python
# adding aditional parameers
plt.plot(dist,speed,marker='*',mec='g',color='b')

plt.show()
```
