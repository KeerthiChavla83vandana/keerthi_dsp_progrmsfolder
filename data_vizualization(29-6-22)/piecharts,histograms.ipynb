```python
import matplotlib.pyplot as plt
import numpy as np
country=['india','bhutan','usa','uk','russia']
Gdp_per_capita=[30000,5000,44000,45000,18000]
#ploting bar graph

plt.bar(country,Gdp_per_capita)

plt.title('country vs gdp')

#country is on x axis name,gdp is values for bar
plt.xlabel('counrty')
plt.ylabel('gdp_per_capita')

# show bar
plt.show()
```

```python
# horizantal bar graph
import matplotlib.pyplot as plt
import numpy as np
plt.title("county vs gdp")

country=['india','bhutan','usa','uk','russia']
Gdp_per_capita=[30000,5000,44000,45000,18000]
plt.xlabel('country')

plt.ylabel('gdp')
#horizontal bar,with color

plt.barh(country,Gdp_per_capita,color='gray')
plt.show()
```

```python
# ploting HISTOGRAM
number=[0.1,0.5,1,1.5,2.4,5.5,6,8,9]

plt.hist(number)
plt.xlabel('number')
plt.ylabel('frequency')

plt.show()
```

```python
# with number of bins
marks=[0.2,0.5,1,1.5,2.4,5.5,6,8,9]

plt.hist(number,bins=5,color='hotpink')
plt.xlabel('marks')
plt.ylabel('frequency')

plt.title('marks vs freqency')

plt.show()
```

```python
mid_marks=[3,4,5,6,8,9,10,11,12,13,14,15]

plt.hist(mid_marks,bins=3,color='yellow') #3-7,7-11,11-15 range with 3 bins

plt.xlabel('sales')
plt.ylabel('no of sales')

plt.title("sales")

plt.show()
           
```

```python
import matplotlib.pyplot as plt
import numpy as np

# drawing home
# MAKIG RECTANGLE POINTS
plt.figure(figsize=(10,10))

plt.xlim(1,10)
plt.ylim(1,10)

plt.plot(2,2,  marker='o',mec='yellow')
plt.plot(8,2, marker='o',mec='yellow')
plt.plot(8,5, marker='o', mec='yellow')
plt.plot(2,5, marker='o', mec='yellow')

# walls for rectangle
plt.plot([2,8,8,2,2],[2,2,5,5,2])

# Triangle
plt.plot(5,7, marker='o', mec='r')

plt.plot([2,5,8],[5,7,5] ,mec='black')

# lines inside of rectangle
plt.plot(3,2 )
plt.plot(3,5 )
plt.plot([3,3],[2,5],mec='b')

plt.plot(4,2 )
plt.plot(4,5 )
plt.plot([4,4],[2,5],mec='b')

plt.plot(5,2 )
plt.plot(5,5 )
plt.plot([5,5],[2,5],mec='b')

plt.plot(6,2 )
plt.plot(6,5 )
plt.plot([6,6],[2,5],mec='b')

plt.plot(7,2)
plt.plot(7,5 )
plt.plot([7,7],[2,5],mec='b')

# line inside od triangle
plt.plot(3,5.5)
plt.plot(7,5.5)
plt.plot([3,7],[5.5,5.5])

plt.plot(4,6.5)
plt.plot(6,6.5)
plt.plot([4,6],[6.5,6.5])



plt.show()
```
