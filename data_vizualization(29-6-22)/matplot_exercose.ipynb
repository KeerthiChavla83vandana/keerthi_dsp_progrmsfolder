```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df=pd.read_csv("company_sales.csv")
#for giving x,y values
profitlist=df["total_profit"].tolist()
monthlist=df["month_number"]

plt.xlabel("month numbers")
plt.ylabel("profit in dollars")

plt.title("month wise profits")



plt.plot(monthlist,profitlist,label='month wise profits data of last year')

plt.xticks(monthlist)
plt.yticks([100000,200000,300000,400000,500000])

plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("company_sales.csv")

profitlist=df['total_profit'].tolist()
monthlist=df['month_number'].tolist()

plt.plot(monthlist,profitlist,label='profit data of last year',color='r',marker='o',linestyle='--',linewidth=3)
plt.xlabel('month number')
plt.ylabel('profit in dollar')
plt.legend(loc='lower right')

plt.title('company sales data of last year')
plt.xticks(monthlist)
plt.yticks([100000,200000,300000,400000,500000])

plt.show()

```

```python
#markerface COLOR
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("company_sales.csv")

profitlist=df['total_profit'].tolist()
monthlist=df['month_number'].tolist()

plt.plot(monthlist,profitlist,label="month wise profits data",color='r',marker='o',markerfacecolor='g',linewidth=3,linestyle='--')
plt.xlabel('month number')
plt.ylabel('profit in dollar')
plt.legend(loc='lower right')

plt.title('company sales data of last year')
plt.xticks(monthlist)
plt.yticks([100000,200000,300000,400000,500000])

plt.show()
```
