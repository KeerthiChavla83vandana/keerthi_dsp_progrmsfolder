## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.4])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python

ax.plot(x,z)
ax.set_xlabel("x")
ax.set_ylabel("z")

#inside plot
ax2.plot(x,y)
ax2.set_xlabel("x")
ax2.set_ylabel("y")

ax2.set_title("zoom")
ax2.set_xlim(20,22) #setting x cords
ax2.set_ylim(30,50) # set y cords

fig


```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig ,sub= plt.subplots(nrows=1,ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
sub[0].plot(x,y,color="blue",ls="--",linewidth=3)

#col 2
sub[1].plot(x,z,color="red",linewidth=3,ls="-")
fig
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig,sub=plt.subplots(nrows=1,ncols=2,figsize=(12,2))

sub[0].plot(x,y,color="blue",lw=5)
sub[0].set_xlabel("x")
sub[0].set_ylabel("y")

sub[1].plot(x,y,color="red",lw=3,ls="--")
sub[1].set_xlabel("x")
sub[1].set_ylabel("z")
```
