```python
#separate plot for each item
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("company_sales.csv")

df=pd.read_csv('company_sales.csv')
monthlist=df['month_number'].tolist()

faceCremSales_data=df['facecream'].tolist()
faceWashSales_data=df['facewash'].tolist()
toothpastSales_data=df['toothpaste'].tolist()
bathingsoapSales_data=df['bathingsoap'].tolist()
shampooSales_data=df['shampoo'].tolist()
moisturizersales_data=df["moisturizer"].tolist()

plt.plot(monthlist,faceCremSales_data,label="face cream Sales Data",marker='o',linewidth=3)
plt.plot(monthlist,faceWashSales_data,label="face wash sales Data",marker='o',linewidth=3)
plt.plot(monthlist,toothpastSales_data,label="toothpaste sales Data",marker='o',linewidth=3)

plt.plot(monthlist,bathingsoapSales_data,label="bathingsoap sales Data",marker='o',linewidth=3)
plt.plot(monthlist,shampooSales_data,label="shampoo Sales Data",marker='o',linewidth=3)
plt.plot(monthlist,moisturizersales_data,label="moisturiser sales Data",marker='o',linewidth=3)

plt.xlabel('month number')
plt.ylabel('sales units in number')
plt.legend(loc="upper left")
plt.xticks(monthlist)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.title('sales data')
plt.show()





```

```python
#ex 8
#pie chart shows no of units sold per year for each product in PERCENTAGE

import pandas as pd
import matplotlib.pyplot as plt

monthlist = df['month_number'].tolist()  #

labels = ['facecream','facewash','Toothpaste','Bathing soap','sShampoo','Moisturizer']

salesData = [df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
#sales data used to dived into slices for dif products andd their percentage

# autopct ENABLES YOU TO DISPLAY PERCENT VALUE USING pYTHON STRING FORMATTING 

plt.axis("equal")

plt.pie(salesData,labels=labels,autopct='%1.1f%%')
print("3d pie cart")



plt.legend(loc='lower right')
plt.title('sales data')
plt.show()
```

```python
#ex 8
#pie chart shows no of units sold per year for each product in PERCENTAGE

import pandas as pd
import matplotlib.pyplot as plt

monthlist = df['month_number'].tolist()  #

labels = ['facecream','facewash','Toothpaste','Bathing soap','sShampoo','Moisturizer']

salesData = [df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
#sales data used to dived into slices for dif products andd their percentage

# autopct ENABLES YOU TO DISPLAY PERCENT VALUE USING pYTHON STRING FORMATTING 

plt.axis("equal")
paints=['lime','r','y','g','violet','b','hotpink']

plt.pie(salesData,labels=labels,autopct='%1.3f%%',colors=paints)
print("3d pie cart")



plt.legend(loc='lower right')
plt.title('sales data')
plt.show()
```

```python
#ex 10 
# Read all products Sales and show it STACK PLOT
print("STACK PLOT")

monthlist = df['month_number'].tolist()
faceCreamSalesData = df['facecream'].tolist()
faceWashSalesData = df['facewash'].tolist()
toothpasteSalesData = df['toothpaste'].tolist()
bathingsoapSalesData=df['bathingsoap'].tolist()
shampooSalesData = df['shampoo'].tolist()
moisturizerSalesData = df['moisturizer'].tolist()

plt.plot([],[],color='m', label= 'face Cream',linewidth=5 )
plt.plot([],[],color='c',label='face wash',linewidth=5)
plt.plot([],[],color='r',label='Tooth paste',linewidth=5)#plt.plot([],[],color='k',label='Bathing soap',linewidth=5)
plt.plot([],[],color='g',label='shampoo',linewidth=5)
plt.plot([],[],color='y',label='Moisturizer',linewidth=5)

plt.stackplot(monthlist,faceCreamSalesData,faceWashSalesData,toothpasteSalesData,bathingsoapSalesData,shampooSalesData,moisturizerSalesData,
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

plt.title('ALL proudcts sales data using stack plot')

plt.legend(loc='upper left')
plt.show()
```
